```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de r (de 1 até 4 com passo de 0.01)
r_vals = np.linspace(1, 4, 3000)
x = 0.5 * np.ones_like(r_vals)

# Número de iterações
n_iter = 1000     # Iterações iniciais para "estabilizar"
n_plot = 1000     # Iterações que serão plotadas após estabilizar

# Lista para armazenar os pontos (r, x)
r_plot = []
x_plot = []

# Iterar para estabilizar
for _ in range(n_iter):
    x = r_vals * x * (1 - x)

# Iterar e armazenar pontos após estabilização
for _ in range(n_plot):
    x = r_vals * x * (1 - x)
    r_plot.extend(r_vals)
    x_plot.extend(x)

# Plotando
plt.figure(figsize=(10, 6))
plt.plot(r_plot, x_plot, ',k', alpha=0.25)
plt.xlabel("r")
plt.ylabel("x")
plt.title("Feigenbaum plot (mapa logístico)")
plt.grid(True)
plt.show()
```

a) Como aparecem no gráfico:

Ponto fixo: Aparece como um único ponto para um dado valor de r (todos os x convergem para o mesmo valor)

Ciclo limite: Aparece como vários pontos discretos para um dado r (x alterna entre um conjunto fixo de valores)

Caos: Aparece como uma "nuvem" densa de pontos aparentemente aleatórios para um dado r

b) Transição para o caos:

O caos começa por volta de r≈3.57, conhecido como borda do caos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuração geral dos plots
plt.figure(figsize=(18, 5))

# ------------------------------------------------------------
# a) Curva Deltoide
# ------------------------------------------------------------
plt.subplot(1, 3, 1)

theta = np.linspace(0, 2*np.pi, 1000)
x = 2*np.cos(theta) + np.cos(2*theta)
y = 2*np.sin(theta) - np.sin(2*theta)

plt.plot(x, y, 'b-', linewidth=2)
plt.title('Curva Deltoide')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')

# ------------------------------------------------------------
# b) Espiral Galileana (r = θ²)
# ------------------------------------------------------------
plt.subplot(1, 3, 2)

theta = np.linspace(0, 10*np.pi, 1000)
r = theta**2
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y, 'r-', linewidth=1)
plt.title('Espiral Galileana (r = θ²)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')

# ------------------------------------------------------------
# c) Função de Fey
# ------------------------------------------------------------
plt.subplot(1, 3, 3)

theta = np.linspace(0, 24*np.pi, 5000)
r = np.exp(np.cos(theta)) - 2*np.cos(4*theta) + np.sin(theta/12)**5
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y, 'g-', linewidth=1)
plt.title('Função de Fey')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')

# Ajusta o layout e mostra os plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import os

# Lista de arquivos das regiões
arquivos = [
    'Norte.xlsx',
    'Nordeste.xlsx',
    'Sudeste.xlsx',
    'Sul.xlsx',
    'Centro-Oeste.xlsx'
]

# Lê e concatena os dados
lista_dfs = []
for arquivo in arquivos:
    if os.path.exists(arquivo):
        df = pd.read_excel(arquivo, header=1)
        df['Região'] = arquivo.replace('.xlsx', '')
        df.drop(columns=['Unnamed: 1', "p"], inplace=True)
        df.dropna(inplace=True)
        lista_dfs.append(df)
    else:
        print(f"Arquivo não encontrado: {arquivo}")

# Exemplo de um dos dados tratados
lista_dfs[0]
```

```python
# Extrai os valores da linha 'Total' para '1º trimestre 2025'
valores_total = {}

for df in lista_dfs:
    regiao = df['Região'].iloc[0]

    # Corrige possíveis nomes de colunas com espaços
    colunas_corrigidas = [col.strip() for col in df.columns]
    df.columns = colunas_corrigidas

    # Filtra linha 'Total' com mais flexibilidade
    linha_total = df[df.iloc[:, 0].astype(str).str.strip().str.lower() == 'total']

    if not linha_total.empty and '1º trimestre 2025' in df.columns:
        valor_bruto = linha_total['1º trimestre 2025'].values[0]

        try:
            # Troca vírgula por ponto se necessário e converte para float
            valor = float(str(valor_bruto).replace(',', '.'))
            valores_total[regiao] = valor
        except ValueError:
            print(f"Valor inválido para a região: {regiao}")
    else:
        print(f"Dados incompletos para a região: {regiao}")

# Converte para listas (na ordem original dos dados)
regioes = list(valores_total.keys())
valores = list(valores_total.values())


# Plota gráfico com eixo x crescente
plt.figure(figsize=(10, 6))
plt.barh(regioes, valores, color='blue')
plt.xlabel('Taxa de Desocupação (%)')
plt.title('Taxa de Desocupação (Total) - 1º Trimestre de 2025')
plt.xlim(left=0)  # Eixo x começa do zero real
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import requests
from io import BytesIO

# 1. Criar DataFrame com seus dados REAIS 
data = {
    'Região': ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste'],
    'Valor': valores  # Valores devem ser numéricos
}
df_regioes = pd.DataFrame(data)

# 2. Carregar shapefile simplificado do Brasil
try:
    url = "https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson"
    brasil = gpd.read_file(url)
    
    # Mapeamento de estados para regiões
    regiao_map = {
        'Norte': ['AC', 'AM', 'AP', 'PA', 'RO', 'RR', 'TO'],
        'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
        'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
        'Sul': ['PR', 'RS', 'SC'],  # Corrigido para capital 'Sul'
        'Centro-Oeste': ['DF', 'GO', 'MS', 'MT']
    }
    
    # Adicionar coluna de região ao shapefile
    brasil['Região'] = ''
    for regiao, estados in regiao_map.items():
        brasil.loc[brasil['sigla'].isin(estados), 'Região'] = regiao
    
    # 3. Juntar dados - verifique correspondência de nomes
    # Converter para o mesmo tipo (maiúsculas/minúsculas)
    df_regioes['Região'] = df_regioes['Região'].str.capitalize()
    brasil['Região'] = brasil['Região'].str.capitalize()
    
    # 4. Dissolver geometrias por região (sem tentar calcular média)
    brasil = brasil.dissolve(by='Região')
    
    # Juntar com dados numéricos
    brasil = brasil.merge(df_regioes, on='Região', how='left')
    
    # 5. Criar o mapa
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    
    # Verificar se temos dados válidos
    if brasil['Valor'].isnull().any():
        print("Aviso: Algumas regiões não têm valores associados")
    
    brasil.plot(column='Valor', ax=ax, legend=True,
               legend_kwds={'label': "Valor por Região", 'shrink': 0.6},
               cmap='Blues', edgecolor='black', linewidth=0.8,
               missing_kwds={'color': 'lightgrey'})
    
    # Adicionar rótulos
    for idx, row in brasil.iterrows():
        if pd.notnull(row['Valor']):
            centroid = row['geometry'].centroid
            ax.annotate(text=f"{row['Região']}\n{row['Valor']:.1f}", 
                        xy=(centroid.x, centroid.y),
                        ha='center', fontsize=10, color='black',
                        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.7))
    
    ax.set_title('Taxa de desocupação por Região do Brasil', fontsize=16)
    ax.axis('off')
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Erro ao criar o mapa: {str(e)}")
    print("Verifique:")
    print("- Sua conexão com a internet")
    print("- Se os nomes das regiões correspondem exatamente")
    print("- Se todos os valores são numéricos")
```

```python
import matplotlib.pyplot as plt

valores_por_regiao = {}
cores = ['skyblue', 'lightgreen', 'lightcoral', 'plum', 'gold']

for df in lista_dfs:
    regiao = df['Região'].iloc[0]
    df.columns = [col.strip() for col in df.columns]
    linha_total = df[df.iloc[:, 0].astype(str).str.strip().str.lower() == 'total']

    if not linha_total.empty:
        colunas_trimestres = [col for col in df.columns if 'trimestre' in col.lower()]
        valores2 = linha_total[colunas_trimestres].values.flatten()
        valores2 = [float(str(v).replace(',', '.')) for v in valores2 if pd.notna(v)]
        valores_por_regiao[regiao] = valores2
    else:
        print(f"Dados da linha 'Total' não encontrados para a região: {regiao}")

# Cria o boxplot com regiões no eixo x e taxa no eixo y
plt.figure(figsize=(10, 6))
box = plt.boxplot(valores_por_regiao.values(),
                  labels=valores_por_regiao.keys(),
                  patch_artist=True,
                  medianprops=dict(color='black'))

# Aplica cores diferentes a cada box
for patch, cor in zip(box['boxes'], cores):
    patch.set_facecolor(cor)
    patch.set_edgecolor('black')

# Rótulos e título
plt.ylabel('Taxa de Desocupação (%)')
plt.title('Distribuição Trimestral da Taxa de Desocupação por Região - Boxplot')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cria df_concat a partir da lista_dfs existente
df_concat = pd.concat(lista_dfs)  # Concatena todos os DataFrames da lista

# Filtra apenas a linha do Total
df_serie = df_concat[df_concat.iloc[:, 0] == 'Total'].drop(columns=['Unnamed: 0'])

# Transforma de wide para long
df_melted = df_serie.melt(id_vars='Região', var_name='Trimestre', value_name='Taxa')

# Trata as colunas
df_melted = df_melted[df_melted['Trimestre'].notna()]
df_melted['Trimestre'] = df_melted['Trimestre'].astype(str)
df_melted['Taxa'] = df_melted['Taxa'].astype(str).str.replace(',', '.')
df_melted['Taxa'] = pd.to_numeric(df_melted['Taxa'], errors='coerce')

# Ordena os trimestres cronologicamente (opcional)
df_melted['Trimestre'] = pd.Categorical(
    df_melted['Trimestre'], 
    ordered=True, 
    categories=sorted(df_melted['Trimestre'].unique(), key=lambda x: (x[-4:], x[:2]))
)

# Plota
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x='Trimestre', y='Taxa', hue='Região', marker='o')
plt.title('Evolução temporal da taxa de desocupação total')
plt.xticks(rotation=90)
plt.ylabel('Taxa (%)')
plt.tight_layout()
plt.grid(True)
plt.show()
```

# Discussão e Análise Crítica:

A taxa de desemprego no Brasil revela profundas desigualdades regionais, que refletem não apenas diferenças econômicas atuais, mas também heranças históricas mal resolvidas, desigualdade estrutural e políticas públicas mal distribuídas. Hoje, observa-se que o Nordeste lidera com a maior taxa de desocupação, seguido pelas regiões Norte, Sudeste, Centro-Oeste e, por último, o Sul, que apresenta o menor índice. Essa ordem, do maior para o menor, varia entre 9,8% e 4,2%, o que demonstra um abismo considerável entre as regiões brasileiras no que se refere ao acesso a oportunidades formais e dignas de trabalho.

As causas dessa disparidade são diversas e multifatoriais. No caso do Nordeste e do Norte, a menor industrialização, os baixos níveis de investimento público e privado e a infraestrutura precária ainda limitam fortemente a criação de empregos sustentáveis. Além disso, a informalidade é mais presente nessas regiões, o que dificulta tanto o acesso a direitos trabalhistas quanto a construção de uma carreira profissional estável. Já no Sul e em partes do Centro-Oeste, a presença de atividades econômicas mais consolidadas, como o agronegócio e setores industriais mais robustos, contribui para uma inserção mais constante, formal e qualificada no mercado de trabalho.

No cenário histórico recente, a taxa de desemprego teve uma trajetória preocupante. A partir de 2017 até o início de 2020, houve um aumento substancial, resultado direto da lenta recuperação da crise econômica de 2014–2016. Essa tendência foi intensificada com os impactos da pandemia da COVID-19, que desorganizou o mercado de trabalho, afetando especialmente os mais vulneráveis. No entanto, atualmente os índices retornaram a patamares próximos aos registrados há cerca de 13 anos, sugerindo uma relativa estabilização. Contudo, essa estabilidade não necessariamente indica uma recuperação de qualidade: parte dela está relacionada à informalidade crescente, ao desalento e à precarização dos vínculos empregatícios em diversas regiões.

Esses dados, por sua vez, devem ser analisados com cautela, considerando as limitações da base de informações utilizadas. A PNAD Contínua, principal fonte para medição do desemprego no Brasil, apesar de ampla e confiável, possui lacunas importantes. Por se basear em entrevistas domiciliares, pode subestimar fenômenos como o trabalho informal e o subemprego, situações em que o indivíduo já não busca mais trabalho por descrença na melhoria do cenário. Além disso, ela não consegue capturar com precisão a qualidade do emprego, o que é fundamental para se compreender as reais condições de inserção do trabalhador na economia brasileira atual.
