```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de r (de 1 até 4 com passo de 0.01)
r_vals = np.linspace(1, 4, 3000)
x = 0.5 * np.ones_like(r_vals)

# Número de iterações
n_iter = 1000     # Iterações iniciais para "estabilizar"
n_plot = 1000     # Iterações que serão plotadas após estabilizar

# Lista para armazenar os pontos (r, x)
r_plot = []
x_plot = []

# Iterar para estabilizar
for _ in range(n_iter):
    x = r_vals * x * (1 - x)

# Iterar e armazenar pontos após estabilização
for _ in range(n_plot):
    x = r_vals * x * (1 - x)
    r_plot.extend(r_vals)
    x_plot.extend(x)

# Plotando
plt.figure(figsize=(10, 6))
plt.plot(r_plot, x_plot, ',k', alpha=0.25)
plt.xlabel("r")
plt.ylabel("x")
plt.title("Feigenbaum plot (mapa logístico)")
plt.grid(True)
plt.show()
```

```python
# Gráfico de radar (visualização livre)
from math import pi

# Seleciona até 5 estados para exemplo
estados_exemplo = dados[col_estado].unique()[:5]
df_radar = dados[dados[col_estado].isin(estados_exemplo)].pivot_table(index=col_estado, columns='Região', values=col_valor, aggfunc='sum').fillna(0)

categorias = list(df_radar.columns)
N = len(categorias)

plt.figure(figsize=(8,8))
for i, row in df_radar.iterrows():
    valores = row.values.tolist()
    valores += valores[:1]  # fecha o círculo
    angulos = [n / float(N) * 2 * pi for n in range(N)]
    angulos += angulos[:1]
    plt.polar(angulos, valores, marker='o', label=i)

plt.xticks([n / float(N) * 2 * pi for n in range(N)], categorias)
plt.title('Gráfico de Radar por Estado e Região')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```
