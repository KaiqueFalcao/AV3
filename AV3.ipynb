```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de r (de 1 at 4 com passo de 0.01)
r_vals = np.linspace(1, 4, 3000)
x = 0.5 * np.ones_like(r_vals)

# Nmero de iteraes
n_iter = 1000     # Iteraes iniciais para "estabilizar"
n_plot = 1000     # Iteraes que sero plotadas aps estabilizar

# Lista para armazenar os pontos (r, x)
r_plot = []
x_plot = []

# Iterar para estabilizar
for _ in range(n_iter):
    x = r_vals * x * (1 - x)

# Iterar e armazenar pontos aps estabilizao
for _ in range(n_plot):
    x = r_vals * x * (1 - x)
    r_plot.extend(r_vals)
    x_plot.extend(x)

# Plotando
plt.figure(figsize=(10, 6))
plt.plot(r_plot, x_plot, ',k', alpha=0.25)
plt.xlabel("r")
plt.ylabel("x")
plt.title("Feigenbaum plot (mapa logstico)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parte a: Curva deltoide
theta = np.linspace(0, 2 * np.pi, 1000)
x = 2 * np.cos(theta) + np.cos(2 * theta)
y = 2 * np.sin(theta) - np.sin(2 * theta)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.title("Curva Deltoide")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True)
plt.show()

# Parte b: Espiral Galileana r = 
theta = np.linspace(0, 10 * np.pi, 1000)
r = theta ** 2
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.title("Espiral Galileana (r = )")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True)
plt.show()

# Parte c: Funo de Fey
theta = np.linspace(0, 24 * np.pi, 2000)
r = np.exp(np.cos(theta)) - 2 * np.cos(4 * theta) + np.sin(theta / 12) ** 5
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.title("Funo de Fey")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import os

# Lista de arquivos das regies
arquivos = [
    'Norte.xlsx',
    'Nordeste.xlsx',
    'Sudeste.xlsx',
    'Sul.xlsx',
    'Centro-Oeste.xlsx'
]

# L e concatena os dados
lista_dfs = []
for arquivo in arquivos:
    if os.path.exists(arquivo):
        df = pd.read_excel(arquivo, header=1)
        df['Regio'] = arquivo.replace('.xlsx', '')
        df.drop(columns=['Unnamed: 1', "p"], inplace=True)
        df.dropna(inplace=True)
        lista_dfs.append(df)
    else:
        print(f"Arquivo no encontrado: {arquivo}")
    
lista_dfs[0]
```

```python
# Extrai os valores da linha 'Total' para '1 trimestre 2025'
valores_total = {}

for df in lista_dfs:
    regiao = df['Regio'].iloc[0]

    # Corrige possveis nomes de colunas com espaos
    colunas_corrigidas = [col.strip() for col in df.columns]
    df.columns = colunas_corrigidas

    # Filtra linha 'Total' com mais flexibilidade
    linha_total = df[df.iloc[:, 0].astype(str).str.strip().str.lower() == 'total']

    if not linha_total.empty and '1 trimestre 2025' in df.columns:
        valor_bruto = linha_total['1 trimestre 2025'].values[0]

        try:
            # Troca vrgula por ponto se necessrio e converte para float
            valor = float(str(valor_bruto).replace(',', '.'))
            valores_total[regiao] = valor
        except ValueError:
            print(f"Valor invlido para a regio: {regiao}")
    else:
        print(f"Dados incompletos para a regio: {regiao}")

# Converte para listas (na ordem original dos dados)
regioes = list(valores_total.keys())
valores = list(valores_total.values())

# Plota grfico com eixo x crescente
plt.figure(figsize=(10, 6))
plt.barh(regioes, valores, color='blue')
plt.xlabel('Taxa de Desocupao (%)')
plt.title('Taxa de Desocupao (Total) - 1 Trimestre de 2025')
plt.xlim(left=0)  # Eixo x comea do zero real
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

valores_por_regiao = {}
cores = ['skyblue', 'lightgreen', 'lightcoral', 'plum', 'gold']

for df in lista_dfs:
    regiao = df['Regio'].iloc[0]
    df.columns = [col.strip() for col in df.columns]
    linha_total = df[df.iloc[:, 0].astype(str).str.strip().str.lower() == 'total']

    if not linha_total.empty:
        colunas_trimestres = [col for col in df.columns if 'trimestre' in col.lower()]
        valores = linha_total[colunas_trimestres].values.flatten()
        valores = [float(str(v).replace(',', '.')) for v in valores if pd.notna(v)]
        valores_por_regiao[regiao] = valores
    else:
        print(f"Dados da linha 'Total' no encontrados para a regio: {regiao}")

# Cria o boxplot com regies no eixo x e taxa no eixo y
plt.figure(figsize=(10, 6))
box = plt.boxplot(valores_por_regiao.values(),
                  labels=valores_por_regiao.keys(),
                  patch_artist=True,
                  medianprops=dict(color='black'))

# Aplica cores diferentes a cada box
for patch, cor in zip(box['boxes'], cores):
    patch.set_facecolor(cor)
    patch.set_edgecolor('black')

# Rtulos e ttulo
plt.ylabel('Taxa de Desocupao (%)')
plt.title('Distribuio Trimestral da Taxa de Desocupao por Regio')
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Caminho do shapefile baixado
gdf_mundo = gpd.read_file("caminho/para/ne_110m_admin_0_countries.shp")

# Filtra apenas o Brasil
gdf_brasil = gdf_mundo[gdf_mundo['ADMIN'] == 'Brazil'].copy()

# Dados fictcios por regio (posies geogrficas simplificadas)
dados_regioes = {
    'Norte': {'coords': (-62, -4), 'taxa': 8.2},
    'Nordeste': {'coords': (-40, -10), 'taxa': 10.1},
    'Sudeste': {'coords': (-43, -20), 'taxa': 9.4},
    'Sul': {'coords': (-51, -28), 'taxa': 7.3},
    'Centro-Oeste': {'coords': (-56, -15), 'taxa': 7.9},
}

# Cria GeoDataFrame das regies
df_regioes = pd.DataFrame([
    {'Regio': regiao, 'Taxa': dados['taxa'], 'geometry': Point(dados['coords'])}
    for regiao, dados in dados_regioes.items()
])
gdf_regioes = gpd.GeoDataFrame(df_regioes, geometry='geometry', crs="EPSG:4326")

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
gdf_brasil.plot(ax=ax, color='white', edgecolor='black')
gdf_regioes.plot(ax=ax, column='Taxa', cmap='OrRd', legend=True, markersize=500)

# Adiciona rtulos
for idx, row in gdf_regioes.iterrows():
    ax.text(row.geometry.x, row.geometry.y + 1.5, row['Regio'], ha='center', fontsize=10, fontweight='bold')

plt.title("Mapa Choropleth  Taxa Total por Regio (1 trimestre 2025)", fontsize=14)
plt.axis('off')
plt.show()
```
