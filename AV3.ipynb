```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de r (de 1 at 4 com passo de 0.01)
r_vals = np.linspace(1, 4, 3000)
x = 0.5 * np.ones_like(r_vals)

# Nmero de iteraes
n_iter = 1000     # Iteraes iniciais para "estabilizar"
n_plot = 1000     # Iteraes que sero plotadas aps estabilizar

# Lista para armazenar os pontos (r, x)
r_plot = []
x_plot = []

# Iterar para estabilizar
for _ in range(n_iter):
    x = r_vals * x * (1 - x)

# Iterar e armazenar pontos aps estabilizao
for _ in range(n_plot):
    x = r_vals * x * (1 - x)
    r_plot.extend(r_vals)
    x_plot.extend(x)

# Plotando
plt.figure(figsize=(10, 6))
plt.plot(r_plot, x_plot, ',k', alpha=0.25)
plt.xlabel("r")
plt.ylabel("x")
plt.title("Feigenbaum plot (mapa logstico)")
plt.grid(True)
plt.show()
```

a) Como aparecem no grfico:

Ponto fixo: Aparece como um nico ponto para um dado valor de r (todos os x convergem para o mesmo valor)

Ciclo limite: Aparece como vrios pontos discretos para um dado r (x alterna entre um conjunto fixo de valores)

Caos: Aparece como uma "nuvem" densa de pontos aparentemente aleatrios para um dado r

b) Transio para o caos:

O caos comea por volta de r3.57, conhecido como borda do caos.

```python
import pandas as pd
import os

# Lista de arquivos das regies
arquivos = [
    'Norte.xlsx',
    'Nordeste.xlsx',
    'Sudeste.xlsx',
    'Sul.xlsx',
    'Centro-Oeste.xlsx'
]

# L e concatena os dados
lista_dfs = []
for arquivo in arquivos:
    if os.path.exists(arquivo):
        df = pd.read_excel(arquivo, header=1)
        df['Regio'] = arquivo.replace('.xlsx', '')
        df.drop(columns=['Unnamed: 1', "p"], inplace=True)
        df.dropna(inplace=True)
        lista_dfs.append(df)
    else:
        print(f"Arquivo no encontrado: {arquivo}")

# Exemplo de um dos dados tratados
lista_dfs[0]
```

```python
# Extrai os valores da linha 'Total' para '1 trimestre 2025'
valores_total = {}

for df in lista_dfs:
    regiao = df['Regio'].iloc[0]

    # Corrige possveis nomes de colunas com espaos
    colunas_corrigidas = [col.strip() for col in df.columns]
    df.columns = colunas_corrigidas

    # Filtra linha 'Total' com mais flexibilidade
    linha_total = df[df.iloc[:, 0].astype(str).str.strip().str.lower() == 'total']

    if not linha_total.empty and '1 trimestre 2025' in df.columns:
        valor_bruto = linha_total['1 trimestre 2025'].values[0]

        try:
            # Troca vrgula por ponto se necessrio e converte para float
            valor = float(str(valor_bruto).replace(',', '.'))
            valores_total[regiao] = valor
        except ValueError:
            print(f"Valor invlido para a regio: {regiao}")
    else:
        print(f"Dados incompletos para a regio: {regiao}")

# Converte para listas (na ordem original dos dados)
regioes = list(valores_total.keys())
valores = list(valores_total.values())


# Plota grfico com eixo x crescente
plt.figure(figsize=(10, 6))
plt.barh(regioes, valores, color='blue')
plt.xlabel('Taxa de Desocupao (%)')
plt.title('Taxa de Desocupao (Total) - 1 Trimestre de 2025')
plt.xlim(left=0)  # Eixo x comea do zero real
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import requests
from io import BytesIO

# 1. Criar DataFrame com seus dados REAIS 
data = {
    'Regio': ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste'],
    'Valor': valores  # Valores devem ser numricos
}
df_regioes = pd.DataFrame(data)

# 2. Carregar shapefile simplificado do Brasil
try:
    url = "https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson"
    brasil = gpd.read_file(url)
    
    # Mapeamento de estados para regies
    regiao_map = {
        'Norte': ['AC', 'AM', 'AP', 'PA', 'RO', 'RR', 'TO'],
        'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
        'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
        'Sul': ['PR', 'RS', 'SC'],  # Corrigido para capital 'Sul'
        'Centro-Oeste': ['DF', 'GO', 'MS', 'MT']
    }
    
    # Adicionar coluna de regio ao shapefile
    brasil['Regio'] = ''
    for regiao, estados in regiao_map.items():
        brasil.loc[brasil['sigla'].isin(estados), 'Regio'] = regiao
    
    # 3. Juntar dados - verifique correspondncia de nomes
    # Converter para o mesmo tipo (maisculas/minsculas)
    df_regioes['Regio'] = df_regioes['Regio'].str.capitalize()
    brasil['Regio'] = brasil['Regio'].str.capitalize()
    
    # 4. Dissolver geometrias por regio (sem tentar calcular mdia)
    brasil = brasil.dissolve(by='Regio')
    
    # Juntar com dados numricos
    brasil = brasil.merge(df_regioes, on='Regio', how='left')
    
    # 5. Criar o mapa
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    
    # Verificar se temos dados vlidos
    if brasil['Valor'].isnull().any():
        print("Aviso: Algumas regies no tm valores associados")
    
    brasil.plot(column='Valor', ax=ax, legend=True,
               legend_kwds={'label': "Valor por Regio", 'shrink': 0.6},
               cmap='Blues', edgecolor='black', linewidth=0.8,
               missing_kwds={'color': 'lightgrey'})
    
    # Adicionar rtulos
    for idx, row in brasil.iterrows():
        if pd.notnull(row['Valor']):
            centroid = row['geometry'].centroid
            ax.annotate(text=f"{row['Regio']}\n{row['Valor']:.1f}", 
                        xy=(centroid.x, centroid.y),
                        ha='center', fontsize=10, color='black',
                        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.7))
    
    ax.set_title('Taxa de desocupao por Regio do Brasil', fontsize=16)
    ax.axis('off')
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Erro ao criar o mapa: {str(e)}")
    print("Verifique:")
    print("- Sua conexo com a internet")
    print("- Se os nomes das regies correspondem exatamente")
    print("- Se todos os valores so numricos")
```

```python
import matplotlib.pyplot as plt

valores_por_regiao = {}
cores = ['skyblue', 'lightgreen', 'lightcoral', 'plum', 'gold']

for df in lista_dfs:
    regiao = df['Regio'].iloc[0]
    df.columns = [col.strip() for col in df.columns]
    linha_total = df[df.iloc[:, 0].astype(str).str.strip().str.lower() == 'total']

    if not linha_total.empty:
        colunas_trimestres = [col for col in df.columns if 'trimestre' in col.lower()]
        valores2 = linha_total[colunas_trimestres].values.flatten()
        valores2 = [float(str(v).replace(',', '.')) for v in valores2 if pd.notna(v)]
        valores_por_regiao[regiao] = valores2
    else:
        print(f"Dados da linha 'Total' no encontrados para a regio: {regiao}")

# Cria o boxplot com regies no eixo x e taxa no eixo y
plt.figure(figsize=(10, 6))
box = plt.boxplot(valores_por_regiao.values(),
                  labels=valores_por_regiao.keys(),
                  patch_artist=True,
                  medianprops=dict(color='black'))

# Aplica cores diferentes a cada box
for patch, cor in zip(box['boxes'], cores):
    patch.set_facecolor(cor)
    patch.set_edgecolor('black')

# Rtulos e ttulo
plt.ylabel('Taxa de Desocupao (%)')
plt.title('Distribuio Trimestral da Taxa de Desocupao por Regio - Boxplot')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cria df_concat a partir da lista_dfs existente
df_concat = pd.concat(lista_dfs)  # Concatena todos os DataFrames da lista

# Filtra apenas a linha do Total
df_serie = df_concat[df_concat.iloc[:, 0] == 'Total'].drop(columns=['Unnamed: 0'])

# Transforma de wide para long
df_melted = df_serie.melt(id_vars='Regio', var_name='Trimestre', value_name='Taxa')

# Trata as colunas
df_melted = df_melted[df_melted['Trimestre'].notna()]
df_melted['Trimestre'] = df_melted['Trimestre'].astype(str)
df_melted['Taxa'] = df_melted['Taxa'].astype(str).str.replace(',', '.')
df_melted['Taxa'] = pd.to_numeric(df_melted['Taxa'], errors='coerce')

# Ordena os trimestres cronologicamente (opcional)
df_melted['Trimestre'] = pd.Categorical(
    df_melted['Trimestre'], 
    ordered=True, 
    categories=sorted(df_melted['Trimestre'].unique(), key=lambda x: (x[-4:], x[:2]))
)

# Plota
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x='Trimestre', y='Taxa', hue='Regio', marker='o')
plt.title('Evoluo temporal da taxa de desocupao total')
plt.xticks(rotation=90)
plt.ylabel('Taxa (%)')
plt.tight_layout()
plt.grid(True)
plt.show()
```
